Где, "<b><font color='yellow'>расчёт по теории</font></b>" означает расчёт частоты с использованием формул, найденных в статье "A new calculation for designing multilayer planar spiral inductors":

$$L = \frac{μ_0 N^2 D_{avg} C_1}{2} \left[ ln(\frac{C_2}{ρ}) + C_3 ρ + C_4 ρ^2 \right]$$

где
$$ ρ = \frac{D_{out} - D_{in}}{D_{out} + D_{in}} $$

$ D_{out} = 62 $, внешний размер контура, среднее значение для типичного прямоугольного контура

$ D_{in} = D_{out} - s n - \frac{w}{2} $, внутренний размер контура

$ D_{avg} = \frac{D_{out} - D_{in}}{2}$, усреднённый диаметр

$ C_1 = 1.27 $, $ C_2 = 2.07 $, $ C_3 = 0.18 $, $ C_4 = 0.13 $

$$ C = \frac{π ε l}{ln \frac{s - w}{w}} $$ где _l_ - длина контура

$$ F = \frac{1}{2 π \sqrt{L C}} $$

А "<b><font color='green'>расчёт по частоте</font></b>" означает расчёт частоты с использованием ёмкости контура, рассчитанной из измеренных частоты и индуктивности в соответствии с уравнением

$$ C = \frac{1}{4 π^2 f^2 L} $$
